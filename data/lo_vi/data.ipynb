```python
import io
def get_data(path):
    with io.open(path, 'rt', encoding='utf-8') as file:
        data = [line.strip() for line in file.readlines()]
    return data
```

```python
lo_data = get_data('drive/MyDrive/lo_vi/train.lo')
vi_data = get_data('drive/MyDrive/lo_vi/train.vi')
print(len(lo_data), len(vi_data))
```

```python
import re
any_lo = re.compile(r'[\u0E80-\u0EFF]')
pairs = [(src.replace(' ',''),trg.strip()) for src,trg in zip(lo_data,vi_data) if bool(any_lo.search(src))]
len(pairs)
```

```python
def add_stop(pair):
    return (pair[0] + '.' if pair[0][-1] != '.' else pair[0],
            pair[1] + '.' if pair[1][-1] != '.' else pair[1])

pairs = list(map(add_stop,pairs))
```

```python
from collections import defaultdict
transl_dict = defaultdict(set)
for src,trg in pairs:
    transl_dict[src].add(trg)
len(transl_dict)
```

```python
no_dif_transl = [(src,list(trg)[0]) for src,trg in transl_dict.items() if len(trg)==1]
dif_transl = [(src,trgs) for src,trgs in transl_dict.items() if len(trgs)>1]
print(len(no_dif_transl),len(dif_transl))
```

```python
with io.open('train.lo','wt',encoding='utf-8') as src_file, io.open('train.vi','wt',encoding='utf-8') as trg_file:
    src_file.writelines(src+'\n' for src,_ in no_dif_transl)
    trg_file.writelines(trg+'\n' for _,trg in no_dif_transl)
```

```python
for i,(src,trgs) in enumerate(dif_transl[:45]):
    print(i+1, src)
    for trg in trgs:
        print('*',trg)
    print()
```
